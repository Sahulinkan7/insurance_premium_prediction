```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('insurance.csv')
```

```python
df['sex'].value_counts()
```

```python
df['smoker'].value_counts()
```

```python
df['smoker']=df['smoker'].map({'no':0,'yes':1})
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder=OneHotEncoder()
```

```python
encoder.fit(df[['region']])
```

```python
encodeddata=encoder.transform(df[['region']])
```

```python
encodeddata_df=pd.DataFrame(encodeddata.toarray(),columns=encoder.get_feature_names_out())
```

```python
encodeddata_df
```

```python
newdf=pd.concat([df,encodeddata_df],axis=1)
```

```python
newdf.drop(columns=['region'],axis=1,inplace=True)
```

```python
newdf.drop(columns=['sex'],inplace=True)
```

```python
newdf
```

```python
traindf,testdf=train_test_split(newdf,test_size=0.2,random_state=44)
```

```python
traindf.head()
```

```python
y_train=traindf['expenses']
x_train=traindf.drop(columns=['expenses'],axis=1)
```

```python
x_test=testdf.drop(columns=['expenses'],axis=1)
y_test=testdf['expenses']
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
```

```python
scaled_x_train = scaler.transform(x_train)
```

```python
scaled_x_test = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(scaled_x_train,y_train)
```

```python
lr_predicted_output=lr.predict(scaled_x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,lr_predicted_output)
```

```python
from sklearn.svm import SVR
```

```python
svr=SVR()
svr.fit(scaled_x_train,y_train)
svr_predicted=svr.predict(scaled_x_test)
r2_score(y_test,svr_predicted)
```

```python
from sklearn.linear_model import Ridge
rdg=Ridge()
rdg.fit(scaled_x_train,y_train)
pred_rdg=rdg.predict(scaled_x_test)
r2_score(y_test,pred_rdg)
```

```python
from sklearn.linear_model import ElasticNet
el=ElasticNet()
el.fit(scaled_x_train,y_train)
el_pred=el.predict(scaled_x_test)
r2_score(y_test,el_pred)
```

```python
from sklearn.tree import DecisionTreeRegressor

dtr=DecisionTreeRegressor()
dtr.fit(scaled_x_train,y_train)
dtr_predicted= dtr.predict(scaled_x_test)
r2_score(y_test,dtr_predicted)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor()
rfr.fit(scaled_x_train,y_train)
rfr_predict = rfr.predict(scaled_x_test)
r2_score(rfr_predict,y_test)
```
